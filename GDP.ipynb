```python
#PART 1A
import pandas as pd 
import numpy as np
# Read data from file 'GDP_States.csv' 
df = pd.read_csv("GDP_States.csv") 
# Preview data
df
```

```python
#Remove year 2016-17
df = df[df.Duration != '2016-17']
df
```

```python
#Divide into two-
df_split = np.split(df, [5], axis=0)
df_crore = df_split[0] #Part 1 is GDP in crores
df_perc= df_split[1] #Part 2 is GDP in percent
df_perc
```

```python
#calculate average
average = df_perc.mean() #skipna = True by default, hence not writing
average.sort_values(axis=0, inplace = True)
average.plot.bar(figsize=(20,10), grid = True)
```

```python
#PART 1B
#read all files and merge
import glob

path = r'States' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, encoding = 'ISO-8859-1')
    df = df[['Item','2014-15']]
    df=df.T
    new_header = df.iloc[0]
    df=df[1:]
    df.columns=new_header
    df['State'] = filename.split('-')[1]
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame
```

```python
frame_per_capita = frame.sort_values(by='Per Capita GSDP (Rs.)')
frame_per_capita.plot.bar(x='State', y='Per Capita GSDP (Rs.)', grid = 'True', figsize=(20,10))
```

```python
#Top 5 states- Goa, Sikkim, Puducherry, Haryana, Kerala
#Bottom 5 states - Bihar, Uttar_Pradesh, Manipur, Assam, Jharkhand
ratio = (frame.loc[frame['State'] == 'Goa',['Per Capita GSDP (Rs.)']])/(frame.loc[frame['State'] == 'Goa',['Per Capita GSDP (Rs.)']])
```
